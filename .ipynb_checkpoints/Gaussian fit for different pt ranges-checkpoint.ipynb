```python
h = output['ptfrac_z_jet_gen']
hlist = []
for i in range(h.axes[1].size-1):
    hlist.append(h[sum,i:i+1,:].values())



```

```python
frac_values = h.axes[2].centers
```

```python

for h in hlist:
    h[0] = h[0]/np.sum(h[0])
    plt.plot(h[0])
    plt.show()

```

```python
from scipy.optimize import curve_fit
def gauss(x,  x0, sigma):
    return ((1/(sigma*np.sqrt(2*np.pi)))*np.exp(-(x - x0) ** 2 / (2 * sigma ** 2)))

parameters, covariance = curve_fit(gauss, frac_values, hlist[0][0],bounds = ([1,0.1],[1.5,0.4]))

mean = parameters[0]
sigma = parameters[1]

fit_dist = gauss(frac_values,mean,sigma)
plt.plot(frac_values,fit_dist,'b--',label = 'fit')

plt.plot(frac_values,hlist[0][0], 'r',label = 'data')
plt.legend()
```
